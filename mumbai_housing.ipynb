```python
df_edit = df.join(pd.get_dummies(df.Flat_Type, dtype = int))
```

```python
df_edit2
```

```python
pd.get_dummies(df.Furnishing, dtype = int)
```

```python
df_edit3 = df_edit2.join(pd.get_dummies(df.Furnishing, dtype = int)).drop(['Furnishing'], axis = 1)
```

```python
df_edit4 = df_edit3.join(pd.get_dummies(df.Parking, dtype = int)).drop(['Parking','Flat_Type','Location'], axis = 1)
```

```python
df_final = df_edit4
df_final
```

```python
df_final.info()
```

```python
#Linear Regression Model

from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
Y = df_final.Flat_Price
Y
```

```python
X = df_final.drop(['Flat_Price'], axis = 1)
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y , test_size = 0.2)
```

```python
X_train.shape , Y_train.shape, X_test.shape , Y_test.shape
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
```

```python
print('Coefficients', model.coef_)
print('Intercept', model.intercept_)
print('Mean squared error (MSE); %.2f' % mean_squared_error(Y_test, Y_pred))
print('Coefficient of determination (R^2): %.2f' % r2_score(Y_test, Y_pred))
```

```python
print(model.score(X_test, Y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train, Y_train)
```

```python
forest.score(X_test, Y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    'n_estimators': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000 ],
    'min_samples_split': [2, 4, 6, 8, 10, 12, 14, 16],
    'max_depth': [None, 4, 6, 8, 10, 12, 14]
}

grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring = 'neg_mean_squared_error',
                           return_train_score = True)

grid_search.fit(X_train, Y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(X_train, Y_train)
```
